```python
d_tr_ohe['Income'] = d_tr_ohe['Income'].fillna(0)
```

```python
x_data=d_tr_ohe.drop('Response', axis=1)
y_data=d_tr_ohe['Response']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)
```

```python
print(x_test.shape, y_test.shape)
```

```python

from sklearn.ensemble import RandomForestClassifier
```

```python

rf = RandomForestClassifier()

rf.fit(x_train, y_train)
```

```python
# En training
pred_tr=rf.predict(x_train)

# En testing
pred_ts=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
df_val_tr=pd.DataFrame({'y_train':y_train, 'pred_tr':pred_tr})

# ¿cuántos aciertos?
sum(df_val_tr.y_train==df_val_tr.pred_tr)/len(df_val_tr)
```

```python
accuracy_score(df_val_tr.y_train, df_val_tr.pred_tr, normalize=True)
```

```python
df_val_ts=pd.DataFrame({'y_test':y_test, 'pred_ts':pred_ts})
```

```python
# ¿cuántos aciertos?
accuracy_score(df_val_ts.y_test, df_val_ts.pred_ts, normalize=True)
```

```python
# atención, asume 0.5 como punto de corte
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
sns.set(font_scale=1.2) #  Ajuste tamaño de letra (var global)

ConfusionMatrixDisplay.from_estimator(
    rf, x_train, y_train,
    display_labels=['no','si'],
    cmap='Blues',
    )
```

```python
data_x=d_tr_ohe.drop('Response', axis=1)
data_y=d_tr_ohe['Response']
```

```python
data_x = data_x.values
data_y = data_y.values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Creamos 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 99)
```

```python
rf.fit(x_train, y_train)
```

```python
rf.predict(x_train)
```

```python
pred_probs=rf.predict_proba(x_train)
```

```python
pred_probs
```

```python
y_prob_tr=pred_probs[:,1]
```

```python
y_prob_tr
```

```python
sns.set(font_scale=1.2) #  Ajuste tamaño de letra (var global)

ConfusionMatrixDisplay.from_estimator(
    rf, x_train, y_train,
    display_labels=['no','si'],
    cmap='Blues',
    )
```

```python
y_pred1=rf.predict(x_train)

sns.set(font_scale=1.5)
conf_mat1=pd.crosstab(index=y_train,
                     columns=y_pred1,
                     rownames=['Actual'],
                     colnames=['Pred'],
                     normalize='index')

import seaborn as sns
sns.heatmap(conf_mat1, annot=True, cmap='Blues', fmt='g')
```
